```python
import os

folder_path = '/content/drive/MyDrive/Chopped' 

v_names = os.listdir(folder_path)

v_names = [name for name in v_names if os.path.isfile(os.path.join(folder_path, name))]

#print(file_names)
```

```python
(v_names)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import librosa
import seaborn as sns
energy_v=[]
time_v=[]
for i in range(len(v_names)): 
    
 k=v_names[i]
 audio_file = '/content/drive/MyDrive/Chopped/' +k 
 audio, sr = librosa.load(audio_file)
 e = np.sum(audio**2)
 energy_v.append(e)
 duration = len(audio) / sr
 time_v.append(duration)
```

```python
ml=[]
for i in range(len(energy_v)):
  if energy_v[i] >= 3 and energy_v[i] <= 20:
    ml.append(v_names[i])
```
