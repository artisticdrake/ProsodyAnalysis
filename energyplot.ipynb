```python
import os

folder_path = 'only_voice' 

v_names = os.listdir(folder_path)

v_names = [name for name in v_names if os.path.isfile(os.path.join(folder_path, name))]

#print(file_names)
```

```python
len(v_names)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import librosa
import seaborn as sns
energy_v=[]
time_v=[]
for i in range(len(v_names)): 
    
 k=v_names[i]
 audio_file = "only_voice/"+k 
 audio, sr = librosa.load(audio_file)
 e = np.sum(audio**2)
 energy_v.append(e)
 duration = len(audio) / sr
 time_v.append(duration)
```

```python
folder_path2 = 'voice_music'  

vm_names = os.listdir(folder_path2)

vm_names = [name for name in vm_names if os.path.isfile(os.path.join(folder_path2, name))]
```

```python
len(vm_names)
```

```python
energy_m=[]
time_m=[]
for i in range(len(vm_names)): 
    
 k=vm_names[i]
 audio_file = "voice_music/"+k 
 audio, sr = librosa.load(audio_file)
 e = np.sum(audio**2)
 energy_m.append(e)
 duration = len(audio) / sr
 time_m.append(duration)
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 4))

sns.histplot(energy_v, ax=ax1,color='red')
ax1.set_xlabel('Energy')
ax1.set_ylabel('No.of samples for each bar')
ax1.set_title('Only voice')

sns.histplot(energy_m, ax=ax2)
ax2.set_xlabel('Energy')
ax2.set_ylabel('No.of samples for each bar')
ax2.set_title('Voice + music')

plt.tight_layout()

plt.show()
```

```python
sns.histplot(energy_v, kde=True,color='red',label="Only Voice")
sns.histplot(energy_m, kde=True, label="Voice+Music")
plt.legend()
plt.xlabel('Energy')
plt.ylabel('No.of samples for each bar')
plt.title('Overview of Only Voice and Voice+Music ')
plt.show()
```
